```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_steps_month = pd.DataFrame(fitbit.groupby('Months_encoded').agg({'Steps':'median'}))
median_steps_month.reset_index(inplace=True)
median_steps_month
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_steps_work_weekend = pd.DataFrame(fitbit.groupby('Work_or_Weekend').agg({'Steps':'median'}))
median_steps_work_weekend.reset_index(inplace=True)
display(median_steps_work_weekend, median_steps_month)
```

```python
# plot with matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))

plt.sca(ax1)
plt.bar(median_steps_month['Months_encoded'], median_steps_month['Steps'])
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.title('Average steps per month')
plt.xticks(range(1,13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

plt.sca(ax2)
plt.bar(median_steps_work_weekend['Work_or_Weekend'], median_steps_work_weekend['Steps'])
plt.xlabel('Day Type')
plt.ylabel('Median Steps')
plt.title('Average steps per work week or weekend')
plt.xticks(median_steps_work_weekend['Work_or_Weekend'], ['Work', 'Weekend'])

fig.show()

save_path = r'C:\Users\zluca\OneDrive\LAURA\Ironhack - data analyst\GitHub\week 4\lab-matplotlib-seaborn\your-code\figures\matplotlib_average_steps.png'
plt.savefig(save_path)
```

```python
##plot with seaborn
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# First subplot: Average steps per month
sns.barplot(x='Months_encoded', y='Steps', data=median_steps_month, ax=ax1)
ax1.set_ylabel('Median Steps')
ax1.set_xlabel('Months')
ax1.set_title('Average steps per month')
ax1.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Second subplot: Average steps per work days or weekend
sns.barplot(x='Work_or_Weekend', y='Steps', data=median_steps_work_weekend, ax=ax2)
ax2.set_title('Average steps per work days or weekend')
ax2.set_xlabel('Day Type')
ax2.set_ylabel('Median Steps')
ax2.set_xticklabels(['Work', 'Weekend'])

plt.show()

save_path = r'C:\Users\zluca\OneDrive\LAURA\Ironhack - data analyst\GitHub\week 4\lab-matplotlib-seaborn\your-code\figures\seaborn_average_steps.png'
plt.savefig(save_path)
```

```python
#plot with plotly
fig = sp.make_subplots(rows=2, cols=1, subplot_titles=("Average Steps per Month", "Average Steps per Work Week or Weekend"))

fig.add_trace(go.Bar(x=median_steps_month['Months_encoded'],
                     y=median_steps_month['Steps'],
                     text=median_steps_month['Steps'],
                     name='Average steps per month',
                     textposition='auto'), row=1, col=1)

fig.add_trace(go.Bar(x=median_steps_work_weekend['Work_or_Weekend'],
                     y=median_steps_work_weekend['Steps'],
                     text=median_steps_work_weekend['Steps'],
                     name='Average steps per work week or weekend',
                     textposition='auto'), row=2, col=1)

fig.update_layout(xaxis=dict(title='Month'),
                  xaxis2=dict(title='Day Type'),
                  yaxis=dict(title='Median Steps'),
                  yaxis2=dict(title='Median Steps'),
                  showlegend=False)

fig.update_layout(height=800, width=800)

folder_path = r'C:\Users\zluca\OneDrive\LAURA\Ironhack - data analyst\GitHub\week 4\lab-matplotlib-seaborn\your-code\figures'
file_name = '\plotly_average_steps.html'       
plot_path = folder_path + file_name    

fig.write_html(plot_path)
fig.show()
```

#### How does the bin size affect your plot?

```python
"""
Ages differences are more visible in histogram with higher bin size. 
So higher bin size is good to understand more precisely the data distribution; if however the goal is to show 
the distribution of 'old' and 'young' people (for example), the bin size can be reduced to show the repartition of old/young more effinciently.
So, the selection of bin size depends on what we want to show.
"""
```

#### What does the last plot tell you about the feature `Age`?

```python
"""
Most of the passengers were relatively young (between 22 and 37 years old). The lower fence of the plot and the minimum value are the same,
so very young passengers are still comprised in the IQR(=range between the third and the first quartile). 
Above the upper fence (which is located at 59 years old ), there are outliers (ages values which are outside the IQR). Those values influence
the mean, making it move to the "upper side" of the plot and skirt from the median. 
However, mean and median are still close to each other, so the number of outliers should not be too high.
"""
```

#### What additional information does the last plot provide about feature `Age`?

```python
"""
Just with the seaborn version of the violin plot, it is possible to visualize the age distribution by looking at the width of the violin. 
It is of course wider around the median value, but depending on how the shape is more squashed or more extended, 
we can tell if majority of the value points are closer or further from the median and also in which direction they tend to go further away.
In this case, the plot is pretty squashed, so we can suppose that a good part of the values points are around the median and/or inside the ICR.
By interacting with the plotly plot, we can also see the numerical value for the Kernel Density Estimation (KDE) per age. 
The KDE represents the probability density of the data at different values along the variable's range. 
It shows where data points are likely to be more concentrated (higher density) and where they are less concentrated (lower density). 
"""
```
